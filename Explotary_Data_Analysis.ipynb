#### Getting the data only contain 50 KM and 50 Mile in the year 2020

```python
year_2020 = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020)]
year_2020.head()
```

#### Splitting The Event name column and seperating the country's from the result

##### Spliting the country name

```python
country_names = df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0)
country_names
```

```python
country_names == "USA"
```

#### Combining both the logic and Displaying the result

```python
df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA")]
```

```python
df2 = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA")]
df2.head()
```

#### Displaying the shape of df2 Dataframe

```python
df2.shape
```

#### Removing country name in Event Name column

```python
df2["Event name"].str.split("(").str.get(0)
```

```python
df2["Event name"] = df2["Event name"].str.split("(").str.get(0)
```

```python
df2.head()
```

#### Calculating the Athelet age for 2025

```python
df2["Athelet_Age"] = 2025 - df2["Athlete year of birth"]
df2["Athelet_Age"].head()
```

#### Removing h from Athelet perfomance column

```python
df2["Athlete performance"] = df2["Athlete performance"].str.split('h').str.get(0)
df2.head()
```

#### Droping unwanted columns in the dataset

##### Dropping these columns : Athlete club, Athlete country, Athlete year of birth, Athlete age category

```python
df2 = df2.drop(["Athlete club","Athlete country","Athlete year of birth","Athlete age category"], axis = 1)
```

```python
df2.head()
```

#### Dropping na values from the dataset

```python
df2.isna().sum()
```

```python
df2 = df2.dropna()
```

```python
df2.isna().sum()
```

#### Checking the duplicate values in dataset

```python
df2[df2.duplicated() == True]
```

#### Reseting the Index 

```python
df2.reset_index(drop = True)
```

#### Reframeing the data types

```python
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

#### Renaming the column names

```python
df2 = df2.rename(columns = {'Year of event':'Year',
                 'Event dates':'Race_day',
                 'Event name':'Race_name',
                 'Event distance/length':'Event_distance/length',
                 'Event number of finishers':'Finisher_count',
                 'Athlete performance':'Athlete_performance',
                 'Athlete gender': 'Athlete_gender',
                 'Athlete average speed':'Athlete_average_speed',
                 'Athlete ID':'Athlete_ID',
                 'Athelet_Age':'Athelet_Age'})
```

```python
df2.head()
```
